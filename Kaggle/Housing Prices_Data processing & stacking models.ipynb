```python
#there is a value with year 2207. Converting it to the year of remodelling of the property
train_df.loc[train_df["GarageYrBlt"]==2207,"GarageYrBlt"] = train_df.loc[train_df["GarageYrBlt"]==2207]["YearRemodAdd"]
```

```python
#create age columns based on year built and year of remodelling
year = datetime.date.today().year
train_df["age"] = year-train_df["YearBuilt"]
train_df["ageAftRemod"] = year-train_df["YearRemodAdd"]
train_df["Garage_Age"] = year - train_df["GarageYrBlt"]
```

```python
#define a dataframe to hold the year and month data
saledate_df = train_df[["YrSold","MoSold"]]

saledate_df = saledate_df.rename(columns={"YrSold": "Year", "MoSold": "Month"})

#Get a date field in the dataframe
from datetime import datetime
saledate_df = saledate_df[["Year","Month"]].assign(Date=pd.to_datetime(saledate_df[["Year","Month"]].assign(day=1)))

#get sale age in years
train_df["sale_age"] = (((datetime.today() - saledate_df["Date"])/np.timedelta64(1, 'M')).astype(int))/12
```

```python
def get_rmse(y_test_pred,y_test,is_log):
    if(is_log == 1):
        rmse = metrics.mean_squared_error(np.floor(np.expm1(y_test)),np.floor(np.expm1(y_test_pred)),squared=False)
    else:
        rmse = metrics.mean_squared_error(y_test, np.floor(y_test_pred),squared=False)
    print(rmse)
```

```python
def get_rmsle(y_test_pred,y_test,is_log):
    if(is_log == 1):
        rmsle = metrics.mean_squared_error(y_test,y_test_pred,squared=False)
    else:
        rmsle =math.sqrt(metrics.mean_squared_log_error(y_test, np.floor(y_test_pred)))
    print(rmsle)
```

```python
def get_rmsle_cv(model,X,y,kfold):
    scores = np.sqrt(-cross_val_score(model,X,y,cv=kfold,scoring="neg_mean_squared_error"))
    print(model.__class__.__name__+": {:.4f} ({:.4f})\n".format(scores.mean(), scores.std()))
```

```python
def create_submission(model,test,filename):
    print('Predict submission at:', datetime.now())
    submission = pd.read_csv("../input/home-data-for-ml-course/sample_submission.csv")
    prediction = np.floor(np.expm1(model.predict(test)))
    submission.iloc[:, 1] = prediction
    submission.to_csv(filename, index=False)
    print("submission_saved at:",datetime.now())
    return prediction
```

```python
from sklearn.pipeline import make_pipeline
#getting all the models
rcv_alphas = [9,10,11]
lcv_alphas = [0.0004,0.0005,0.0006]
e_alphas = [0.0009,0.001,0.002]
e_l1ratio = [0.4,0.5,0.6]


lr_final = LinearRegression()
rcv_final = RidgeCV(alphas=rcv_alphas,cv=kfold_10)
lcv_final = LassoCV(alphas = lcv_alphas,cv=kfold_10,random_state=1001)
encv_final = ElasticNetCV(max_iter=1e7,alphas=e_alphas,l1_ratio=e_l1ratio, cv=kfold_10,random_state=1001)
dtr_final = tree.DecisionTreeRegressor(min_samples_split=22,min_samples_leaf=7,max_features='auto',
                                       max_depth=20,random_state=1001)
rfr_final = RandomForestRegressor(n_estimators=300,min_samples_split=5,min_samples_leaf=1,max_features='auto',
                                max_depth=20,bootstrap=True,random_state=1001)
lgbm_final = LGBMRegressor(learning_rate=0.006,n_estimators=2500,max_depth=4,min_child_samples=3,num_leaves=62,random_state=1001)
xgb_final = XGBRegressor(learning_rate=0.03,n_estimators=1000,max_depth=4,min_child_weight=3,gamma=0,
                         colsample_bytree=0.4,subsample=0.6,random_state=1001)
gbr_final = GradientBoostingRegressor(learning_rate=0.05,n_estimators=2100,max_depth=7,min_samples_leaf=6,
                                    min_samples_split=800,max_features='sqrt',random_state = 1001)
svr_final = SVR(C=10000,epsilon=0.01,gamma=1e-06)
knn_final = KNeighborsRegressor(leaf_size=1,n_neighbors=7,p=2)
krr_final = KernelRidge(alpha=1,coef0=7,degree=1)
```
